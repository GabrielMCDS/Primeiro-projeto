```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Crop Yeild Data.csv')
```

```python
#Informações gerais sobre o dataset.
def check_dataframe(df):
    print(f"{' HEAD ':-^50}")
    print(df.head())

    print(f"{' DESCRIBE ':-^50}")
    print(df.describe())

    print(f"{' INFO ':-^50}")
    print(df.info())

    print(f"{' TYPES ':-^50}")
    print(df.dtypes)

    print(f"{' SHAPE ':-^50}")
    print(df.shape)


check_dataframe(df)

'''Crop: The name of the crop (e.g., Rice, Wheat, Coconut).
Crop_Year: Year of harvest.
Season: Season of planting (Kharif, Rabi, etc.).
State: Indian State.
Area: Farm size in Hectares.
Production: Total production in Tonnes.
Annual_Rainfall: Annual rainfall in mm.
Fertilizer: The total amount of fertilizer used for the crop (in kilograms).
Pesticide: The total amount of pesticide used for the crop (in kilograms).
Avg_Temperature (NEW): Average annual temperature (°C) fetched from NASA.
Max_Temperature (NEW): Maximum annual temperature (°C).
Min_Temperature (NEW): Minimum annual temperature (°C).
Yield: Production / Area (Tonnes/Hectare).'''
```
